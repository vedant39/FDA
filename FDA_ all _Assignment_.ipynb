**One-Sample t-Test**

```python
from scipy.stats import ttest_ind
```

```python

from scipy.stats import ttest_1samp

null_mean = 100000

sample_data = df['Annual Salary']

t_statistic, p_value = ttest_1samp(sample_data, null_mean)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < 0.05:
  print("Reject the null hypothesis: The average annual salary is significantly different from $100,000.")
else:
  print("Fail to reject the null hypothesis: The average annual salary is not significantly different from $100,000.")
```

**Two-Sample t-Test:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

import numpy as np
from scipy.stats import ttest_ind

null_mean = 0

sample_data1 = df[df['Gender'] == 'Male']['Annual Salary']
sample_data2 = df[df['Gender'] == 'Female']['Annual Salary']

t_statistic, p_value = ttest_ind(sample_data1, sample_data2)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < 0.05:
  print("Reject the null hypothesis: The average annual salary is significantly different between genders.")
else:
  print("Fail to reject the null hypothesis: The average annual salary is not significantly different between genders.")
```

### **Paired t-Test**

```python
from scipy.stats import ttest_rel
```

```python

null_mean = 0

sample_data1 = df['Annual Salary'].iloc[:10]
sample_data2 = df['Annual Salary'].iloc[10:20]

t_statistic, p_value = ttest_rel(sample_data1, sample_data2)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < 0.05:
  print("Reject the null hypothesis: The average annual salary is significantly different between the two groups.")
else:
  print("Fail to reject the null hypothesis: The average annual salary is not significantly different between the two groups.")
```

## **ANOVA (F-Test)**

```python

import numpy as np
import pandas as pd
from scipy import stats

df = pd.DataFrame({
    'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'value': [5, 6, 7, 8, 9, 10, 11, 12, 13]
})

data = df.groupby('group')['value'].apply(list)

f_statistic, p_value = stats.f_oneway(*data)

print("F-statistic:", f_statistic)
print("p-value:", p_value)

if p_value < 0.05:
  print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
  print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

```python

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = df['Age']
y = df['Annual Salary']
z = df['Bonus %']

x_line = np.linspace(min(x), max(x), 100)
y_line = np.linspace(min(y), max(y), 100)
z_line = np.linspace(min(z), max(z), 100)

ax.plot3D(x_line, y_line, z_line, color='red')

ax.scatter(x, y, z, c='blue', marker='o')

ax.set_xlabel('Age')
ax.set_ylabel('Annual Salary')
ax.set_zlabel('Bonus %')

plt.title('3D Plot of Age, Annual Salary, and Bonus % with Line Plot')
plt.show()
```

```python

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = df['Age']
y = df['Annual Salary']
z = df['Bonus %']

x_surf, y_surf = np.meshgrid(np.linspace(min(x), max(x), 100), np.linspace(min(y), max(y), 100))
z_surf = np.zeros_like(x_surf)

for i in range(len(x_surf)):
  for j in range(len(y_surf)):
    z_surf[i][j] = df[(df['Age'] >= x_surf[i][j]) & (df['Annual Salary'] >= y_surf[i][j])]['Bonus %'].mean()

ax.plot_surface(x_surf, y_surf, z_surf, cmap='viridis', alpha=0.8)

ax.scatter(x, y, z, c='blue', marker='o')

ax.set_xlabel('Age')
ax.set_ylabel('Annual Salary')
ax.set_zlabel('Bonus %')

plt.title('3D Surface Plot of Age, Annual Salary, and Bonus %')
plt.show()
```
